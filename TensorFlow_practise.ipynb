```python
import tensorflow as tf
```

```python
scaler  = tf.constant(7)
vector = tf.constant([1,3])
matrix = tf.constant([[1,2],[3,4]])

print("Scaler : ",scaler.shape,scaler.dtype)
print("vector : ",vector.shape,vector.dtype)
print("matrix : ",matrix.shape,matrix.dtype)
```

```python
#proof that network only likes numbers
numbers = tf.constant([1,2])
weights = tf.constant([3,4])

print("number dtype :",numbers.dtype)
print("math on number :",(numbers*weights).numpy())

words = tf.constant(["I","Love","Pizza"])
print("words dtype :",words.dtype)

try:
  result = words*weights
  print(result)
except Exception as e:
  print("Math on words fails --> ",  type(e).__name__)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

sentences = [
    "Neural networks are inspired by the human brain.",
    "Machine learning helps computers learn from data.",
    "Python is one of the most popular programming languages.",
    "Data preprocessing is an important step in machine learning.",
    "Gradient descent is used to minimize the loss function.",
    "Activation functions introduce non-linearity into neural networks.",
    "Deep learning models require large amounts of data.",
    "Cross-entropy loss is commonly used for classification tasks.",
    "Regularization helps prevent overfitting in machine learning models.",
    "Transformers have revolutionized natural language processing."
]

tokenizer = Tokenizer()

tokenizer.fit_on_texts(sentences)

print("Token found and their count :")
print(tokenizer.word_counts)
print()
print('Number of unique tokens ', len(tokenizer.word_counts))

#word_index :
print("Word Index (token --> ID)")
print(tokenizer.word_index)

#Now convert each sentense into the list of Integer IDs
sequences = tokenizer.texts_to_sequences(sentences)

print()
for seq,origin in zip(sequences,sentences):
  print(seq,"<---",origin)

#here we observed that all length of words are different
print()
for s in sequences:
 print(s)

print("Padding Only !!!!")

print("\n POST padding (Zeroes at the end): ")
print(pad_sequences(sequences,padding='post'))

print("\n PRE padding (Zeroes at the starting) ---> This is by default: ")
print(pad_sequences(sequences,maxlen=5,truncating='pre',padding='pre'))


print("Padding with truncating !!!!")

print("\n POST padding (Zeroes at the end): and truncating is post --> cut at the END")
print(pad_sequences(sequences,maxlen=5,truncating='pre',padding='post'))

print("\n PRE padding (Zeroes at the starting) ---> and truncating is post --> cut at the STARTING")
print(pad_sequences(sequences,maxlen=5,truncating='pre',padding='pre'))
```
